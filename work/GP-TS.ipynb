```python
import numpy as np
import matplotlib.pyplot as plt

import BanditProblems as bp
```

```python
number_of_action = 1000
number_of_arms   = 100
train_func = bp.train_one_peak
bandit_algo = bp.gp_ts
x = np.linspace(0, 1, number_of_arms)
y = train_func(x)[1]
bp.plot_train_func(x,y)
```

```python
T = 1
sum = 0
kernel_sigma  = [5, 0.01]
for i in range(T):
    y_train, x_train = bp.sim_GP(bandit_algo, number_of_action, train_func, x, bp.GP, is_gp_plot=True, kernel_sigma=kernel_sigma)
    print(f'sum reward = {np.sum(y_train)}')
    sum += np.sum(y_train)
print(f'average sum_reward = {sum/T}')
```

```python
train_area = {'mean' : 0.5, 'min' : 0.4, 'max' : 0.6},
bp.plot_transition_action(x_train, number_of_action, train_area)
bp.plot_sum_reward(y_train)
```
