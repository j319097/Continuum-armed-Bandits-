```python
# import sys
# sys.path.append('..') 
import numpy as np
import matplotlib.pyplot as plt

import BanditProblems as bp
```

```python
number_of_action = 1000
number_of_arms   = 1000
train_func = bp.train_one_peak
bandit_algo = bp.gp_ts
x = np.linspace(0, 1, number_of_arms)
y = train_func(x)[1]
train_area = {'mean' : 0.5, 'min' : 0.4, 'max' : 0.6},
```

```python
bp.plot_train_func(x,y)
```

```python
puls_arm_numbers = [1, 3, 5, 10, 30, 50]
T = 30
for p_arm in puls_arm_numbers:
    sum = 0
    for i in range(T):
        sum_r, x_train = bp.sim_TS(number_of_action, train_func, x, bp.TS, puls_arm_number = p_arm)
        sum += sum_r
        if i == 1:
            title = f' puls_arm_number = {p_arm}: total reward = {sum_r}: percent = {(p_arm*3)/number_of_arms} '
            bp.plot_transition_action(x_train, number_of_action, train_area, title)
    print(f'average sum_reward = {sum/T}')
```
